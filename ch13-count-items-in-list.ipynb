### create a dictionary through a for loop

```python
list = [102, 232, 424]
count = 0
d = {} #Empty dictionary to add values into

for i in list:
    d[count] = i
    count+=1
d
```

```python
list = [102, 232, 424]
d = {}

for a, b in enumerate(list):
    d[a] = b

d
```

```python
d = {a:b for a, b in enumerate(list)}
d
```

```python
dict(zip(range(0, len(list)), list))
```

### LIST ITEMS TO A DICTIONARY

```python
items = "Whats the simpliest way to add the list items to a dictionary "

stats = {}
for i in items:
    if i in stats:
        stats[i] += 1
        print(stats)
    else:
        stats[i] = 1
        print("\t",stats)

# bonus
# for i in sorted(stats, key=stats.get):
#    print("%d√ó'%s'" % (stats[i], i))
```
