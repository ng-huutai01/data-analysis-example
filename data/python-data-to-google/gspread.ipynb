### Finding a Cell


```python
# Find a cell matching a string:
cell = wsheet.find("tphcm")

print("Found something at R%sC%s" % (cell.row, cell.col))
```

```python
# Find a cell matching a regular expression
import re
amount_re = re.compile(r'(cổ truyền|Không quân)')
cell = wsheet.find(amount_re)
cell # kq: R19C2 nghĩa là dòng 19 cột số 2
```

### Finding All Matched Cells

```python
# Find all cells matching a string:
cell_list = wsheet.findall("Hà Nội")
cell_list
```

```python
# Find all cells matching a regexp:
criteria_re = re.compile(r'(phòng khám|Bệnh viện) đa khoa')
cell_list = wsheet.findall(criteria_re)
cell_list
```

### FORMATING

```python
# Set A1:B1 text format to bold:
wsheet.format('A2:B2', {'textFormat': {'bold': True}})
```

```python
# Color the background of A2:B2 cell range in black, change horizontal alignment, text color and font size:
wsheet.format("A2:B2", {
    "backgroundColor": {
      "red": 0.0,
      "green": 0.0,
      "blue": 0.0
    },
    "horizontalAlignment": "CENTER",
    "textFormat": {
      "foregroundColor": {
        "red": 1.0,
        "green": 1.0,
        "blue": 1.0
      },
      "fontSize": 12,
      "bold": True
    }
})

# A full specification of format options is available at CellFormat in Sheet API Reference.
# https://developers.google.com/sheets/api/reference/rest/v4/spreadsheets/cells#cellformat

# gspread-formatting offers extensive functionality to help you when you go beyond basics.
# https://github.com/robin900/gspread-formatting
```

### Using gspread with pandas

```python
import pandas as pd
dataframe = pd.DataFrame(wsheet.get_all_records())
dataframe
```

__writing a dataframe to a sheet. With update() we put the header of a dataframe into the first row of a sheet followed by the values of a dataframe:__

```python
wsheet.update([dataframe.columns.values.tolist()] + dataframe.values.tolist())
```

### Using gspread with NumPy


```python
# The code above assumes that your data starts from the first row of the sheet. 
# If you have a header row in the first row, you need replace worksheet.get_all_values() with worksheet.get_all_values()[1:].
import numpy as np
array = np.array(wsheet.get_all_values())
```

__Write a NumPy array to a sheet:__

```python
array = np.array([[1, 2, 3], [4, 5, 6]])

# Write the array to worksheet starting from the A2 cell
wsheet.update('A2', array.tolist())
```
